```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
import bqplot
import numpy as np 
```

```python
df=pd.read_csv('state_employee.csv',na_values={'YTD Gross': 0, 'Period Pay Rate': 0})
df['Agency Code'] = pd.factorize(df['Agency'])[0]
df['Division Code'] = pd.factorize(df['Agency Division'])[0]
df.loc[df['YTD Gross']<=15800, 'Income Level'] = 0
df.loc[(df['YTD Gross']>15800)&(df['YTD Gross']<=20500), 'Income Level'] = 1
df.loc[(df['YTD Gross']>20500)&(df['YTD Gross']<=26500),'Income Level'] = 2
df.loc[df['YTD Gross']>26500, 'Income Level'] = 3
df
#The Column of Agency was transferred into Agency Code(0-72)
#The Column of Agency Division was transferred into Division Code(0-965)
#The Column of YTD Gross was transferred into Income Level. The standard is below:
# YTD Gross<=15800 --- 0,which represents Low
# 15800<YTD Gross<=20500 --- 1,which represents Medium Low
# 20500<YTD Gross<=26500 --- 2,which represents Medium High
# YTD Gross>26500 --- 3,which represents High
```

```python
data_map = df.groupby(['Agency Code','Income Level']).agg({'YTD Gross':'mean'}).reset_index()
data_map = data_map.pivot('Agency Code', 'Income Level', 'YTD Gross')
data_map=data_map.fillna(0)
data_map
```

```python
income = df.groupby(by=['YTD Gross']).agg({'Income Level': 'first'}).reset_index()
income
```

```python
data_line = df.pivot_table(values='YTD Gross', index=['Income Level','Agency Code','Division Code'], aggfunc='mean').reset_index()
data_line
```

```python
mySelectedLabel = ipywidgets.Label()

#heat map
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
col_sc = bqplot.ColorScale(scheme = "Reds")

c_axmap = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', side = 'right')
x_axmap = bqplot.Axis(scale = x_sc,label='Income Level')
y_axmap = bqplot.Axis(scale = y_sc,label='Agency Code', orientation = 'vertical')

heat_map = bqplot.GridHeatMap(color = data_map,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'}, 
                              anchor_style = {'fill':'blue'}, 
                              selected_style = {'opacity': 1.0}, 
                              unselected_style = {'opacity': 0.8}) 

#line plot
# Set up x/y scales
ax_sc = bqplot.LinearScale()
ay_sc = bqplot.LinearScale()

# Set p axis
x_axline = bqplot.Axis(scale = ax_sc, label='Agency Division Code')
y_axline = bqplot.Axis(scale = ay_sc, 
                   orientation = 'vertical', 
                   label='Average YTD Gross(USD)')

#(1) set up marks
lines = bqplot.Scatter(x = data_line['Division Code'],
                     y = data_line['YTD Gross'],
                     scales = {'x': ax_sc, 'y': ay_sc})


def get_data_value(change):
    if len(change['owner'].selected) ==1:
        i,j = change['owner'].selected[0]
        agency = df.groupby(by=['Agency']).agg({'Agency Code': 'first'}).reset_index()
        income = df.groupby(by=['YTD Gross']).agg({'Income Level': 'first'}).reset_index()
        region_mask = (data_line['Agency Code'] == i) & (data_line['Income Level']==j) & (data_line['Division Code']>=-1)
        lines.x = data_line['Division Code'][region_mask]
        lines.y = data_line['YTD Gross'][region_mask]

heat_map.observe(get_data_value, 'selected')

fig_heatmap = bqplot.Figure(marks = [heat_map], axes = [c_axmap, y_axmap, x_axmap])
fig_line = bqplot.Figure(marks = [lines], axes = [x_axline, y_axline])

fig_heatmap.layout.min_width='400px'
fig_line.layout.min_width='500px'

myDashboard = ipywidgets.VBox([ipywidgets.HBox([fig_heatmap,fig_line])])
myDashboard
```

A good way to make use of this dataset is to explore the average YTD Gross according to the agency and its division.
I divided the YTD Gross for the data points into 4 levels.YTD Gross<=15800 is 0,which represents low income; 
15800<YTD Gross<=20500 is 1,which represents medium low income; 20500<YTD Gross<=26500 is 2,which represents medium high income; YTD Gross>26500 is 3, which represents high income. The heatmap is to locate the agency and the income level, then the scatter plot would display the average YTD Gross for the income level under this agency and the agency division. It helps us thoroughly know the income distribution for each agency and its division from low to high levels. 


The Agency and Agency Division are very important. Since the dashboard cannot display the complicated text labels, the code was generated to represents the text content. However, we should know the original text information otherwise the code will be meaningless.
